```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Brand': ['Tata', 'Tesla', 'Audi', 'BMW'],
    'Location': ['India', 'US', 'Germany', 'Germany'],
    'Model':['Ace', '3', 'Q3', 'Z4'],
    'Sales': [100,75,84,92]
})
df
```

```python
pd.pivot(data=df, index='Brand', columns='Location', values=['Sales']).fillna(0)
```

```python
df.pivot(columns='Location', values=['Model', 'Sales'], index='Brand')
```

```python
tesla_s = pd.Series(['Tesla', 'US', 'S', 40], index=['Brand', 'Location', 'Model', 'Sales'])
tesla_s
```

```python
df = df.append(tesla_s, ignore_index=True)
df
```

```python
df.pivot(columns='Location', values=['Model', 'Sales'], index='Brand')
```

```python
df.pivot_table(index='Brand', columns='Location', values=['Model', 'Sales'],
               aggfunc={'Model':'count', 'Sales':'mean'})
```

```python
df.pivot_table(index=['Brand', 'Model'], columns='Location', values=['Sales'],
               aggfunc={'Sales':'mean'})
```

```python
b = pd.read_csv('data/AirBnb_NYC_2019.csv', nrows=10000)
b
```
