## merge

```python
'Adam Robin Mike Andrew Jack Aaron'.split(' ')
```

```python
eng = pd.DataFrame({'name_1':'Adam Robin Mike Andrew Jack Aaron'.split(' '),
                     'eng':[50,41,24,18,30,21]})
eng
```

```python
math = pd.DataFrame({'name_2':['Adam', 'Adam', 'Dennis', 'Aaron', 'Andrew', 'Jack', 'Mark'],
                     'math':[10,15,42,70,2,22,11]})
math
```

```python
# inner merge

pd.merge(left=eng, right=math, how='inner', left_on='name_1', right_on='name_2', indicator=True)
```

```python
# outer merge

pd.merge(left=eng, right=math, how='outer', left_on='name_1', right_on='name_2', indicator='True')
```

```python
# left merge

pd.merge(left=eng, right=math, how='left', left_on='name_1', right_on='name_2', indicator='qwerty')
```

```python
# right merge

pd.merge(left=eng, right=math, how='right', left_on='name_1', right_on='name_2', indicator=True)
```

```python
eng1 = pd.DataFrame({'fname_1':'Adam Robin Mike Andrew Jack Aaron'.split(' '),
                     'lname_1': 'Smith Johnson Ramsey Silver Rice Bould'.split(' '),
                     'eng':[50,41,24,18,30,21]})
eng1
```

```python
math1 = pd.DataFrame({'fname_2':['Adam', 'Adam', 'Dennis', 'Aaron', 'Andrew', 'Jack', 'Mark'],
                     'lname_2':['Smith', 'Smith', 'Lillee', 'Lee', 'Jackson', 'Morrison', 'Coy'],
                     'math':[10,15,42,70,2,22,11]})
math1
```

```python
# multi index merge

pd.merge(left=eng1, right=math1, how='left', left_on=['fname_1', 'lname_1'],
         right_on=['fname_2', 'lname_2'], indicator=True)
```

## join

```python
sci = pd.DataFrame({'name':'Wayne Frank Pat Tony Lee Sol'.split(' '),
                     'sci':[41,51,87,71,87,47]})
sci.set_index('name', drop=True, inplace=True)
sci
```

```python
geo = pd.DataFrame({'name':['Rocky', 'Wayne', 'Leo', 'Frank', 'Andre', 'Tony', 'Sol'],
                     'geo':[55,17,41,28,88,24,90]})
geo.set_index('name', drop=True, inplace=True)
geo
```

```python
sci2 = pd.DataFrame({'name':['Rocky', 'Jason', 'Leo', 'Michael', 'Andre', 'Tony', 'Sol'],
                     'sci2':[55,17,41,28,88,24,90]})
sci2.set_index('name', drop=True, inplace=True)
sci2
```

```python
sci.join(geo, how='inner')
```

```python
geo2 = geo.reset_index()
geo2
```

```python
# in the tutorial video, I showed how the lsuffix, rsuffix arguments work using 'sci' dataframe,
# but that required manipulating the original sci dataframe which would mess with other code cells.
# hence here I've changed the sci dataframe to geo, to demonstrate how and when the l/r suffix
# arguments are needed.
geo2.join(geo, on='name', how='inner', lsuffix='_left')
```

```python
# geo, sci, sci2

geo.join([sci, sci2], how='inner')
```

## TASKS:

1. Obtain the same result as above (with the merge and how='inner') with how='inner' using join.

```python
# Here you'll see how merge and join work interchangeably 
```

```python
# method 1: Making both dataframe columns to be joined upon as the indexes:

c.set_index('Country', inplace=True)
h.set_index('Country (region)', inplace=True)
```

```python
c.head()
```

```python
h.head()
```

```python
# in the pd.merge operation, we had used happiness (h) dataframe as the left dataframe,
# and we'll continue in the same vein here too:

h.join(c, how='inner')

# you see? it is the same result as obtained using pd.merge. You'll notice before, using pd.merge,
# we had 15 columns and now in this it is only 13 columns. Can you figure out why?

# it's because, in this join, both the dataframes' country columns are indexes, and hence removed as
# normal columns. Whereas in merge, both the columns were regular columns.
```

```python
# method 2:

# having only one dataframe's column to be merged upon as the index.

# we'll make happiness (h) dataframe as normal again with country as a regular column:

h.reset_index(inplace=True)
```

```python
h.head()
```

```python
# now we know that the df that has the joining column as its index, should be placed inside the ():

h.join(c, on='Country (region)', how='inner')
```

#### Bonus tip:

You know that if you have both columns to be joined upon are indexes of their respective dataframes, then join is the way to go.

But what if you still want to use pd.merge?

pd.merge provides this functionality through left_index and right_index parameters

If we set both these parameters to True, pandas will automatically look for the index columns of both these dataframes and infer results, which is basically what join does in the first place, right!

The covid dataframe already has its joining column as the index. But the happiness dataframe does not. Hence we'll do that now:

```python
h.set_index('Country (region)', inplace=True)
```

```python
pd.merge(left=h.set_index('Country (region)'), right=c, left_index=True, right_index=True)
```

2. There are two datasets, namely spotify-top30.csv and spotify-30-50.csv in the data folder, containing the list of songs most listened on spotify in a particular year. Glue these two datasets together to make a top 50 compilation

```python
spotify1 = pd.read_csv('data/spotify-top30.csv')
spotify1.head()
```

```python
spotify2 = pd.read_csv('data/spotify-30-50.csv')
spotify2.head()
```

```python
print("Spotify top 30 df shape:", spotify1.shape)
print("Spotify 31-50 df shape:", spotify2.shape)
```

### spotify2 has two less columns than spotify1, and now we'll see how this alters our answers:

### since we want to make a compilation of top 50 songs, it'd make sense to glue these dataframes one below the other, right?

```python
pd.concat([spotify1, spotify2], sort=False, join='outer') # outer is the default join type
```

You'd have noticed that 'Beats.Per.Minute' and 'Energy' columns are missing from the spotify2 dataset, hence those columns have been filled with NaNs for the 31-50 ranked songs

```python
# if we do join='inner':

pd.concat([spotify1, spotify2], sort=False, join='inner')

# now  'Beats.Per.Minute' and 'Energy' columns have been removed, as we asked for an inner join!
```
