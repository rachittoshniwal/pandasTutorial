4. Filter out rows from AirBnb NYC data set where the name of the host contains the letter 'L', maybe because I MIGHT have seen an astrologer and he MIGHT have suggested that people whose name contains 'L' keep their rooms clean, or maybe its just  me who wants to test your row-filtering skills... who knows?

```python
nyc.info()
```

```python
# First,  we'll check the host_name column to see who all contain the letter 'l' in their names.

# We do this by first 'stringifying' the column using .str attribute and then calling .contains
# attribute off it and passing the string we want to match. We also have to take care of whether
# lower and upper case matters in our case. If not, then pass case=False.

# Also, as we can see above, the host_name column contains a few missing (NaN) values. Pandas won't
# know what to do with them, and hence we have to step up and tell it that whenever you encounter a
# NaN value, replace it with False.

nyc['host_name'].str.contains('l', case=False, na=False)
```

```python
# and now we'll make a dataframe off these boolean values

nyc[nyc['host_name'].str.contains('l', case=False, na=False)]
```

6. I've just won a lottery(!) and I now have too much money at my disposal. I've bought a brand new Android phone with the latest specs, and now I want to buy each and every paid app available in the store. Pls tell me how much it'll cost me, not that it really matters because I've just won a lottery!

```python
# first we'll see the boolean series of our need
store['Type'] == 'Paid'
```

```python
# now, we'll make a dataframe off it

store[store['Type'] == 'Paid']
```

```python
# Now since we're only interested in the price column, we'll filter that out:

price = store[store['Type'] == 'Paid']['Price']
price
```

```python
# now as you can see this series is an 'object' i.e. a string because it contains the dollar symbol
# We will get rid of this using the split() method which'll split each row according to our criteria

price_lst = []
for each_row in price:
    price_lst.append(pd.to_numeric(each_row.split('$')[1]))
price_lst
```

```python
# Was that too much? I'll break it down for you.
# First we initialize an empty list: price_lst = []
# Next we loop over each row in the price column, splitting each of those rows on the '$' value
# and then take the second element of the splitted row
```

```python
# for eg, the 234th row contains an app that is paid, and it is priced at $4.99
store['Price'][234]
```

```python
# as we can see here, the split makes a list of the values after splitting on our given string ($)
store['Price'][234].split('$')
```

```python
# now we're only interested in the second element of this list, i.e. 4.99. Hence:

store['Price'][234].split('$')[1]
```

```python
# Now this 4.99 is a string, expemplified by the quotes around it, or we can check it using type()

type(store['Price'][234].split('$')[1])
```

```python
# Now we need to convert it to a numeric value, we'll do it using pd.to_numeric()

# and now that everything has been converted into a numeric, we can call the sum() method
# after converting it into a series:

pd.Series(price_lst).sum()

# It'll cost me over $11136 to buy out all the apps in the play store!
```
